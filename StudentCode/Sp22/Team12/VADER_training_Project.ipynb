```python
from operator import neg
from nltk.tag import pos_tag
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
nltk.download('vader_lexicon')
import pandas as pd
df = pd.read_csv("combined.csv").iloc[650000:970000,:]
sentences = df["text"]
target = df["target"]

total = len(sentences)
correct = 0
pos_total = 0
neg_total = 0
pos_correct = 0
false_pos = 0
false_neg = 0
neg_correct = 0


sid = SentimentIntensityAnalyzer()
df['comp_scores'] = df["text"].apply(lambda sentence: sid.polarity_scores(str(sentence))["compound"])
df['simple'] = df["comp_scores"].apply(lambda score: "positive" if score >= 0 else 'negative')
df["target_word"] = df['target'].apply(lambda target: "positive" if target >= 0 else 'negative')


print("accuracy score: ")
print(accuracy_score(df['target_word'],df['simple']))

print(classification_report(df['target_word'],df['simple']))


  
```
